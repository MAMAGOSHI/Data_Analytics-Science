```python
#Write a function that takes a row as input and returns a shortened version of the values in the Variety
#column. The values should consist of the first two words of the variety name, followed by a space and an
#ellipsis (…). Hint: You’ll need to split the values into a list of words, get the first two words in the list, add
#the ellipsis to the list, and then join the words in the list.

def shorten_variety(row):
    words = row.split()[:2]  
    return ' '.join(words) + ' ...'
```

```python
#Apply the function you just wrote to the DataFrame, and store the results in a column named
#ShortVariety. Then, display the first five rows to be sure this worked.

ramen_ratings['ShortVariety'] = ramen_ratings['Variety'].apply(shorten_variety)

print(ramen_ratings[['Variety', 'ShortVariety']].head())
```
